```python
base = pd.DataFrame()  # Start with an empty DataFrame  
  
directory = 'county_gdp'  
for count,csv_file in enumerate(os.listdir(directory)):  
    df = pd.read_csv(os.path.join(directory, csv_file), index_col=None).drop(columns=['Unnamed: 1'])
    county_name = csv_file.split('_')[0]
    df.rename(columns={"2001":f"{county_name}2001",
                    "2002":f"{county_name}2002",
                    "2003":f"{county_name}2003",
                    "2004":f"{county_name}2004",
                    "2005":f"{county_name}2005",
                    "2006":f"{county_name}2006",
                    "2007":f"{county_name}2007",
                    "2008":f"{county_name}2008",
                    "2009":f"{county_name}2009",
                    "2010":f"{county_name}2010",
                    "2011":f"{county_name}2011",
                    "2012":f"{county_name}2012",
                    "2013":f"{county_name}2013",
                    "2014":f"{county_name}2014",
                    "2015":f"{county_name}2015",
                    "2016":f"{county_name}2016",
                    "2017":f"{county_name}2017",
                    "2018":f"{county_name}2018",
                    "2019":f"{county_name}2019",
                    "2020":f"{county_name}2020",
                    "2021":f"{county_name}2021",
                    "Unnamed: 0": "CountyYear"}, inplace=True)
    df=df.transpose()
    if count != 0:
        df = df.iloc[1:,:]
    base = pd.concat([base, df], axis=0)
        
base = base.iloc[0:, :]
# After concatenation, save this large DataFrame to a CSV  
base.to_csv('combined_county_gdp.csv')  
```
