```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
ROOT = os.getcwd()
print(f'Root directory: {ROOT}')
```

```python
# check population estimate data
population_estimate_path = 'PopEstimates2000-2022.csv'
population_estimate = pd.read_csv(population_estimate_path)
population_estimate
```

```python
# check gdp data
gdp_path = 'combined_county_gdp.csv'
gdp = pd.read_csv(gdp_path)
gdp
```

```python
# put population estimate data in the same format as gdp data
count=0
for county,year in zip(population_estimate['County'],population_estimate['Year']):
    county = county.split(' ')[0]
    new_row = f"{county}{year}"
    population_estimate.loc[count,'CountyYear'] = new_row
    count+=1
population_estimate.drop(columns=['County','Year'],inplace=True)
#move CountyYear to the first column
population_estimate = population_estimate.loc[:,['CountyYear','Pop']]
population_estimate
```

```python
# merge population estimate and gdp data on CountyYear
# perform a left merge to keep all the rows in population estimate
merged_data = pd.merge(population_estimate,gdp,on='CountyYear',how='left')
merged_data
```

```python
merged_data.to_csv('utah_countys_pop_gdp.csv',index=False)
```
